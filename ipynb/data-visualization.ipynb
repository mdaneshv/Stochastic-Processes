```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(0,5*np.pi,100)
y = np.cos(x)
z = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y, color='r', label='cos(x)',alpha=0.7)
ax.plot(x,z, color='b', label='sin(x)',alpha=0.7)
ax.set_title('A Simple Plot')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig, ax=plt.subplots(1, figsize=(7,5))
x = [3,2,3,5,3]
y = [2,3,4,3,3]
m = ['o', '>', '<', 'o', '$\clubsuit$']
c = ['r', 'b', 'orchid', 'orange', 'g']
size = [100,100,100,100,300]

for i in range(len(x)):
    ax.scatter(x[i], y[i], marker=m[i], s=size[i], color=c[i], alpha=1)
plt.title('Scatter Plot')    
```

```python
fig,ax = plt.subplots(2)
x = ['green', 'red', 'blue']
y = [10, 2,16]
ax[0].bar(x,y,width=1,label='colors')
ax[1].bar(x,y,label='col')
ax[0].set_title('bar chart')
ax[0].legend(['colors'],fontsize=12)
ax[1].legend(['colors'],fontsize=12)
fig.text(-.01,0.5,'Likes', rotation='vertical', fontsize=13)
plt.tight_layout()
```

```python
plt.figure(figsize=(6,5))

data = [[2,4,1], [1,3,2]]
im = plt.imshow(data, cmap=plt.get_cmap('hot'), interpolation='nearest',
               vmin=-10, vmax=10)
fig.colorbar(im)
plt.title('A heat map')
plt.show()
```

```python
a = np.reshape(np.random.normal(0,1,100),(10,10))
print(a)
plt.imshow(a, plt.get_cmap('hot'))
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.linspace(-10,10,100)
y = -np.linspace(-10,10,100)
X,Y = np.meshgrid(x,y)
Z= np.cos(X**2+Y**2)
ax.contourf(X,Y,Z)
ax.set_title('Contour Plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]
X,Y = np.meshgrid(x,y)
Z = np.exp(X+Y)
plt.contourf(X,Y,Z)
plt.show()
```
